## Functions

```python
'''
Define a function that for a list of item IDs:
    1. Plots a KDE for their values
    2. Outputs a table showing statistical values
    3. If multiple itemids - asseses their similarity to judge whether they can be coalesced
'''
def compare_itemids(df, item_ids):

    # --- Plot a KDE
    plt.figure(figsize = (7, 5))

    #Remove NaNs
    df = df[['itemid', 'valuenum']]
    df.dropna(inplace=True)

    # KDE plots for each groups
    for i in item_ids:
        sns.kdeplot(df.loc[df['itemid'] == i, 'valuenum'], label = i)

    # Labeling of plot
    plt.ylabel('Density');

    plt.show()
    
    # -- Output stats
    stats = (chart_df[chart_df['itemid'].isin(item_ids)].groupby(['dbsource', 'itemid', 'label'])
                                           .agg({'hadm_id': 'nunique',
                                                 'valuenum': ['median', 'std']})
                                           .reset_index())
    
    tot_adm = chart_df.hadm_id.nunique()
    covered_adm = chart_df[chart_df['itemid'].isin(item_ids)].hadm_id.nunique()
    p = (covered_adm / tot_adm) * 100
    print("TOTAL ADMISSIONS COVERED (IF COMBINED): {} out of {} -- {} PERCENT".format(covered_adm, tot_adm, p))
    
    return stats
```

```python
## --- Import all chart events, exclude from the exclude csv, 

def import_chart_events():

    # Import chart events
    chart_df = pd.read_csv(os.path.abspath(os.path.join(os.getcwd(), os.pardir, 'data', 'final_chart_and_lab_events.csv')), index_col=0)

    # Total Admissions
    metavision_tot = chart_df[chart_df['dbsource'] == 'metavision'].hadm_id.nunique()
    carevue_tot = chart_df[chart_df['dbsource'] == 'carevue'].hadm_id.nunique()
    labs_tot = chart_df[chart_df['dbsource'] == 'lab'].hadm_id.nunique()

    # Find % of admissions having readings
    chart_df_s = (chart_df.groupby(['dbsource', 'itemid', 'label'])
                        .agg({'hadm_id':'nunique'})
                        .reset_index()
                        .sort_values(by='hadm_id', ascending=False))

    chart_df_s['perc'] = np.where(chart_df_s['dbsource'] == 'metavision',
                                chart_df_s['hadm_id'] / metavision_tot,
                                np.where(chart_df_s['dbsource'] == 'carevue',
                                chart_df_s['hadm_id'] / carevue_tot,
                                chart_df_s['hadm_id'] / labs_tot))
    return chart_df, chart_df_s
```

```python
'''
Define a function that searches all lab/chart events labels
'''
def search_labels(query):
    results = (chart_df[chart_df['label'].str.contains(query, case=False)].groupby(['dbsource', 'itemid', 'label'])
                                                                          .agg({'hadm_id':'nunique'})
                                                                          .reset_index()
                                                                          .sort_values(by='hadm_id', ascending=False))
    return results
```

## Running

```python
# Fresh import
chart_df, chart_df_s = import_chart_events()
```

```python
# Show only top percentages (to see what events to explore)

# What itemids will I be excluding?
exclude_itemid = pd.read_csv(os.path.abspath(os.path.join(os.getcwd(), os.pardir, 'data', 'exclude_itemid.csv')))
exclude = exclude_itemid.itemid.tolist()

chart_df_s = chart_df_s[(chart_df_s['perc'] > 0.50) & (~chart_df_s['itemid'].isin(exclude))].sort_values(by='perc', ascending=False)
print(len(chart_df_s))
chart_df_s
```
